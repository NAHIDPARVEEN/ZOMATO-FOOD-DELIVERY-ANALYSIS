```python
#  Q.1  Area-wise distribution of restaurant
```

```python
# Calculate the count of restaurants per area
restaurant_per_area = res_detail.groupby('new_location')['id'].count()
print(restaurant_per_area)
```

```python
# Plotting the bar chart using Seaborn
plt.figure(figsize=(16, 4))
sns.barplot(x=restaurant_per_area.index, y=restaurant_per_area, palette='viridis')
plt.title('Area-wise Distribution of Restaurants')
plt.xlabel('Area')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90, ha='right')  # Rotating x-axis labels for better visibility
plt.show()
```
